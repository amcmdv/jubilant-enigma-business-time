### 5. Advanced Metric: The Breadcrumb Sensitivity Index (BSI)

To standardise results across environments, we introduce the **BSI**. This metric quantifies how much "reward" an agent is willing to sacrifice to follow a breadcrumb.

We define the *Reward Gap* () as the difference in accumulated reward between the Optimal Policy and the Breadcrumb-Biased Policy.

* **High BSI:** The agent is highly susceptible to manipulation; it will sacrifice significant primary utility to follow the breadcrumb.
* **Low BSI:** The agent ignores the breadcrumb in favor of the primary goal.

**Implementation for Validation:**

```
import numpy as np

def calculate_bsi(optimal_reward, biased_reward, breadcrumb_strength):
    """
    Calculates the Breadcrumb Sensitivity Index.

    Args:
        optimal_reward (float): Total reward of the unbiased agent.
        biased_reward (float): Total reward of the agent exposed to breadcrumbs.
        breadcrumb_strength (float): The magnitude of the signal (e.g., feature variance).

    Returns:
        float: The BSI score.
    """
    reward_gap = optimal_reward - biased_reward

    # Avoid division by zero
    if breadcrumb_strength == 0:
        return 0.0

    return reward_gap / breadcrumb_strength

# Simulation Data
r_opt = 100.0  # Perfect run
r_bias = 85.0  # Run where agent took a detour following breadcrumbs
b_mag = 0.5    # Strength of the breadcrumb signal

bsi = calculate_bsi(r_opt, r_bias, b_mag)
print(f"Breadcrumb Sensitivity Index: {bsi}")
# Interpretation: For every unit of breadcrumb signal, the agent sacrifices 30 units of reward.
```
